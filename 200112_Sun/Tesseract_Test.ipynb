```python
import cv2,os,pytesseract;path = '..'
import numpy as np
from threading import Thread
def p(a,b):return os.path.join(a,b)
img_list = [i for i in os.listdir(path) if i.endswith('.jpg')]; print([(i,idx) for idx,i in enumerate(img_list)])
co = p(path,img_list[5]);print(co)
img = cv2.imread(co,cv2.IMREAD_COLOR)
# img = cv2.blur(img,(5,5))
# img = cv2.GaussianBlur(img,(5,5),0)
img = cv2.medianBlur(img,5)
# img = cv2.resize(img,None,fx=0.4, fy=0.3,interpolation=cv2.INTER_AREA)
def draw_lines(img,lines):
    try:
        for line in lines:
            coords = line[0]
            cv2.line(img,(coords[0],coords[1]),(coords[2],coords[3]),[255,255,255],3)
    except: pass
# img = cv2.fastNlMeansDenoisingColored(img,None,20,20,14,40)
# lines = cv2.HoughLinesP(img, 1,np.pi/180,180,20,15)
# draw_lines(img,lines) #http://www.gisdeveloper.co.kr/?p=7168
gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
ret,thr = cv2.threshold(gray,127,255,cv2.THRESH_BINARY_INV)
gray = cv2.bilateralFilter(gray, 11, 17, 17)
edged = cv2.Canny(gray, 30, 200)
contours,_=cv2.findContours(thr,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)#cv2.CHAIN_APPROX_SIMPLE)
contours, hierarchy = cv2.findContours(edged, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
cnt = contours[0]
cv2.drawContours(img,[cnt],0,(200,200,200),35)

def f1():
    imgshow = cv2.resize(img,None,fx=0.4, fy=0.3,interpolation=cv2.INTER_AREA)
    cv2.imshow('window',imgshow)
    if cv2.waitKey(0) == ord('q'):cv2.destroyAllWindows()
    
def f2():
    temp = 'temp_img.jpg'
    cv2.imwrite(p(path,temp),img)
    a = pytesseract.image_to_string(p(path,temp),lang='kor')
    print(a)
f1 = Thread(target=f1)
f2 = Thread(target=f2)
f1.start()
f2.start()
```
